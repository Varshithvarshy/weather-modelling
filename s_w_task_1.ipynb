```python
import numpy as np
import matplotlib.pyplot as plt

def temperature_calculate_hardcoded(time):
    a = 0.1
    b = -1
    c = 30

    temperature = a * time ** 2 + b * time + c
    return temperature

def main():
    # Generate time values
    random_values = np.linspace(0, 10, 50)

    # Calculate temperatures using the hardcoded quadratic function
    temperatures = temperature_calculate_hardcoded(random_values)

    # Write data to a file
    filename = "temperature_data.txt"
    with open(filename, 'w') as f:
        f.write("Time (hours)\tTemperature (C)\n")
        for time, temp in zip(random_values, temperatures):
            f.write(f"{time:.2f}\t\t{temp:.2f}\n")

    print(f"Data written to '{filename}' successfully.")

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(random_values, temperatures, label='Temperature')

    plt.xlabel('Time (hours)')
    plt.ylabel('Temperature (C)')
    plt.grid(True)
    plt.title('Quadratic Temperature Model')

    plt.legend()
    plt.savefig('temperature_plot.png')  # Save the plot as an image file
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np


def quadratic_model(time, a, b, c):
  temperature = (a * time * time + b * time + c) * 0.01
  return temperature


list = [(19, 20, 21), (12, 19, 25), (16, 21, 31)]
time_values = np.linspace(0, 12, 50)

for i, (a, b, c) in enumerate(list):
  temperature_values = quadratic_model(time_values, a, b, c)
  label = f'Set{i+1}: a={a},b={b},c={c}'
  plt.plot(time_values, temperature_values, label=label)

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.legend()
plt.grid(True)
plt.title('weather modeling with quadratic equation(multiple set)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * time**2 + b * time + c
    return temperature

a = 0.1
b = -1
c = 30

def temperature_calculation(time):
    a = float(input("Enter the value for a: "))
    b = float(input("Enter the value for b: "))
    c = float(input("Enter the value for c: "))
    temperature = quadratic_model(time, a, b, c)
    return temperature

time_values = np.linspace(0, 10, 50)

temperature_hardcoded = quadratic_model(time_values, a, b, c)

temperature_user = temperature_calculation(time_values)

plt.plot(time_values, temperature_hardcoded, label='Hardcoded Values')
plt.plot(time_values, temperature_user, label='User Input Values')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.grid()
plt.title('Weather Modelling')
plt.show()
```
